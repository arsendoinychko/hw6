```python
#7
t = []
for s in df['Month']:
    if s == "01":
        t = t + [1]
    elif s == "02":
        t = t + [2]
    elif s == "03":
        t = t + [3]
    elif s == "04":
        t = t + [4]
    elif s == "05":
        t = t + [5]
    elif s == "06":
        t = t + [6]
    elif s == "07":
        t = t + [7]
    elif s == "08":
        t = t + [8]
    elif s == "09":
        t = t + [9]
    elif s == "10":
        t = t + [10]
    elif s == "11":
        t = t + [11]
    elif s == "12":
        t = t + [12] 
```

```python
df["Month"] = t
```
